```python
#import torch library,which helps to build and train the model,and handle torch operations
!pip install torch
import torch
import numpy as np
```

```python
data=[
    [1,2],
    [3,4]
]
```

```python
np_array=np.array(data)
x=torch.from_numpy(np_array)
print(type(x))
print(x)
```

```python
x_ones=torch.ones_like(x)
print("One tensor with the same shape of X is:\n",x_ones)
```

```python
x_rand=torch.rand_like(x,dtype=torch.float)
print("Random tensor with the same shape of X is:\n",x_rand)
```

```python
shape=(2,3)
```

```python
random_tensor=torch.rand(shape)
print("Random tensor of above shape is:\n",random_tensor)

ones_tensor=torch.ones(shape)
print("One tensor of above shape is:\n",ones_tensor)

zeros_tensor = torch.zeros(shape)  
print("Zeros tensor of above shape is:\n",zeros_tensor)
```
